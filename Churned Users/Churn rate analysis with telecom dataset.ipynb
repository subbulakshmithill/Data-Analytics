### Indexing and retrieving data

A DataFrame can be indexed in a few different ways such as using column name (label) or row name (index value) or by using a series of index value. 
1. Indexing on a single column: Dataframe['Name']
2. Boolean Indexing: Dataframe[C(Dataframe['Name'])] where C is a condition based on which the results will be outputed.
3. loc() method is used for indexing by name, and iloc() is used for indexing by number.

**Indexing on a single column:**

```python
df['churn'].mean()
```

14.5% is the proportion of churned users in the dataset.

**Boolean Indexing:**

```python
df[df['churn'] == 1].mean()
```

**What is the churned user rate based on the state that they are located?**

```python
df[df['churn'] == 1]['state'].value_counts()
```

```python
sns.countplot(x='state', hue='churn', data=df)
fig=plt.gcf()
fig.set_size_inches(19,8)
```

It can be seen that for some states like TX,NJ,MD,MI,MN,NY,etc., the number of churn users are higher.

```python
sns.lineplot(x="state", y="customer service calls", hue='churn',style="churn",markers=True,dashes=False, data=df)
fig=plt.gcf()
fig.set_size_inches(19,8)
```

It can be seen that in the states which had more churned users like AR(11),AL(8),ID(9),IN(9),etc., the users had to make more customer service calls than the regular users.

```python
columns_to_show = ['voice mail plan','international plan']
df.groupby(['churn'])[columns_to_show].count()
```

```python
pd.crosstab(df['churn'], [df['voice mail plan'],df['international plan']])
```

```python
sns.catplot(x="international plan", hue="churn", col="voice mail plan",
               data=df, kind="count",
                height=4, aspect=.7);
fig=plt.gcf()
fig.set_size_inches(15,8)
```

It can be seen that 
1. Out of 2180 users with no international plan and voice mail plan, about 302 users are churned users i.e approximately 14%. 
2. Among 231 users with international plan and no voice mail plan, 101 users are churned users i.e approximately 44%.
3. Out of 92 users with international plan and voice mail plan, about 36 users are churned users i.e approximately 39%. 
4. Among 830 users with voice mail plan and no international plan, 44 users are churned users i.e approximately 5%.
Hence it can be possible that the international plan or the voice mail plan is not well recieved by the users.
